## 4. Verificar GPU

```python
import torch

print(f'PyTorch: {torch.__version__}')
print(f'CUDA disponible: {torch.cuda.is_available()}')

if torch.cuda.is_available():
    print(f'GPU: {torch.cuda.get_device_name(0)}')
    mem = torch.cuda.get_device_properties(0).total_mem / 1e9
    print(f'Memoria GPU: {mem:.1f} GB')
else:
    print('⚠️ No hay GPU disponible.')
    print('Ve a Runtime > Change runtime type > Hardware accelerator > GPU')
```

## 5. Verificar Estructura del Proyecto

```python
from pathlib import Path

required_dirs = ['configs', 'data', 'src', 'notebooks', 'scripts']
required_files = ['requirements.txt', 'configs/default.yaml', 'src/config.py']

print('Verificando estructura del proyecto...\n')

all_ok = True
for d in required_dirs:
    exists = Path(d).exists()
    status = '✅' if exists else '❌'
    print(f'  {status} {d}/')
    if not exists:
        all_ok = False

for f in required_files:
    exists = Path(f).exists()
    status = '✅' if exists else '❌'
    print(f'  {status} {f}')
    if not exists:
        all_ok = False

print()
if all_ok:
    print('✅ Estructura del proyecto correcta')
else:
    print('⚠️ Faltan algunos archivos o directorios')
```
